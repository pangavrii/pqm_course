
# Python Συναρτήσεις

Παραδείγματα:
- Τι είναι συνάρτηση
- Τι είναι όρισμα
- Τι είναι προεπιλεγμένο όρισμα (default)
- Τι είναι όρισμα λέξης-κλειδί (keyword argument)



## 1 Βασική Συνάρτηση

Θεωρούμε την ενέργεια αρμονικού ταλαντωτή:

$E = \hbar \omega (n + 1/2)$


```python

def energy(n, omega):
    return omega * (n + 0.5)

energy(0, 2)
```


## 2️ Προεπιλεγμένο Όρισμα (Default)

Συχνά βάζουμε ħ = 1 (natural units).


```python

def energy(n, omega, hbar=1):
    return hbar * omega * (n + 0.5)

energy(0, 2)        # χρησιμοποιεί hbar=1
energy(0, 2, 0.5)   # αλλάζουμε το hbar
```


## 3 Ορίσματα Λέξεων-Κλειδιών (Keyword Arguments)

Μπορούμε να δώσουμε όνομα (αυτό που ορίζεται στην συνάρτηση) στα ορίσματα.


```python

energy(0, omega=2, hbar=2)
#energy(omega=2, n=0)
```


## 4️ Συνδυασμός Positional και Keyword


Μπορείς να τα χρησιμοποιήσεις μαζί όταν:

- Τα positional arguments μπαίνουν ΠΡΩΤΑ
- Τα keyword arguments μπαίνουν ΜΕΤΑ

```python

energy(0, omega=2)
```

Δεν  επιτρέπεται να βάλεις positional μετά από Keyword

```python
energy(omega=2,0)
```


## 5 Παράδειγμα Κυματοσυνάρτησης (Gaussian)

$Ψ(x) = exp(-(x-x_0)^2 / (2σ^2))$


```python

import numpy as np

def gaussian(x, x0=0, sigma=1):
    return np.exp(-(x - x0)**2 / (2 * sigma**2))

x = np.linspace(-5, 5, 100)

gaussian(x)              # default

```

```python
gaussian(x, sigma=0.5)   # αλλάζουμε μόνο το sigma
```
